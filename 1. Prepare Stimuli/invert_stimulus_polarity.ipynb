# **Script for inverting Auditory Stimulus Polarity**

## Script Info
---
- This script takes an untriggered mono audio file of unspecified polarity and creates a new mono audio file of the contrary polarity.

- Changing the stimulus polarity is a method used in speech ABR analysis, where one can calculate two seperate sub-averages - one for each polarity, respectively. These sub-averages can then be summed with each other or substracted from one another.

- Make sure to compare your input and output file, making sure that the polarity of your stimulus was indeed inverted. When comparing the two, you should see that the envelope of your soundwave was flipped upside down. When comparing our /da/ syllable file (da_syllable) with the output file of this script (da_syllable_negative), this looks as follows:

> grafik.png

- The usefulness of alternating polarities for your study heavily depends on what part of the ABR you want to analyse. Skoe & Kraus (2010)ยน provide a comprehensive tutorial on the use of stimulus polarities, and other things to consider when conducting speech ABR research.

- ยน Skoe, E., & Kraus, N. (2010). Auditory brainstem response to complex sounds: a tutorial. Ear and hearing, 31(3), 302.

- You can easily adapt this script to your own needs by swapping out the audio files and adapting parameters like trigger pulse width, channel position, etc.

## Invert Soundwave Polarity
---



```python
# Read the wav file (mono)
sampling_rate, data = wavfile.read('da_syllable.wav')                           # Make sure to use your untriggered mono channel files here, as you'll invert your trigger pulse otherwise

# Multiply by -1 to invert the polarity
data_inverted = (data * -1)

# Write the output
wavfile.write('da_syllable_negative.wav', sampling_rate, data_inverted)
files.download('da_syllable_negative.wav')                                      # Only necessary for immediately downloading output to local machine (put '#' in front of code if you dont need it)
```
